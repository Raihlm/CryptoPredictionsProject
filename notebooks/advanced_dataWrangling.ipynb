# Data Cleaning processing dataset that we will use

```python
df_main = pd.read_csv('../Dataset/Historical/HistoricalData.csv')
df_secondary = pd.read_csv('../Dataset/raw/Bitcoin Historical Data From 5 June to 1 Jul.csv')
```

```python
df_main.info()
df_secondary.info()
```

```python
df_main = df_main.drop(columns=['Unnamed: 0'])
```

```python
df_main.describe()
```

```python
df_main.head(5)
```

```python
df_secondary.head(5)
```

```python
df_secondary.columns = df_secondary.columns.astype(str).str.strip().str.lower()
df_secondary = df_secondary.rename(columns={
    'date': 'date',
    'price': 'price',
    'open': 'open',
    'high': 'high',
    'low': 'low',
    'vol.': 'volume',
    'change %': 'change'
})
```

```python
def convert_shorthand(x):
    x = str(x).replace(',', '').replace('%', '').strip()
    if x == '-' or x == '':
        return None
    if x.endswith('K'):
        return float(x[:-1]) * 1_000
    elif x.endswith('M'):
        return float(x[:-1]) * 1_000_000
    elif x.endswith('B'):
        return float(x[:-1]) * 1_000_000_000
    else:
        return float(x)

for col in ['price', 'open', 'high', 'low', 'change']:
    df_secondary[col] = (
        df_secondary[col].astype(str)
               .str.replace(',', '')
               .str.replace('%', '')
               .replace('-', None)
               .astype(float)
    )

df_secondary[col]= df_secondary[col].apply(convert_shorthand)
```

```python
df_secondary['date'] = pd.to_datetime(df_secondary['date'])
df_main['date'] = pd.to_datetime(df_main['date'])
```

```python
df_secondary.head(10)
```

```python
df_secondary.info()
df_main.info()
```

```python
merged_df = pd.concat([df_main, df_secondary])


merged_df = merged_df.drop_duplicates(subset='date')

merged_df = merged_df.sort_values('date')
```

```python
merged_df.to_csv('../Dataset/processed/BitcoinAllHistoricalPrice.csv', index=False)
```
